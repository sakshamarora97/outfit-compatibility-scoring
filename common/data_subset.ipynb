```python
all_train_image_files =[]
for train_outfit in trans_train_outfit_list:
    for img_idx in train_outfit["items_index"]:
        img_file = train_outfit["set_id"] + "_"+ str(img_idx)+".json"
        all_train_image_files.append(img_file)
```

```python
all_test_image_files =[]
for test_outfit in trans_test_outfit_list:
    for img_idx in test_outfit["items_index"]:
        img_file = test_outfit["set_id"] + "_"+ str(img_idx)+".json"
        all_test_image_files.append(img_file)
```

```python
all_image_files = all_train_image_files+all_test_image_files
```

```python
fic = open('data/Backup/category_summarize_100.json', 'r')
category_item = json.load(fic)
```

```python
from collections import defaultdict
new_category_summarize_100 = []
for cat in category_item:
    print(cat['name'])
    cat_summary = defaultdict()    
    cat_summary['id'] = cat['id']
    cat_summary['name'] = cat['name']
    cat_summary_item_list = []
    for item in cat['items']:
        item_file = str(item)+".json"
        if item_file in all_image_files:
            cat_summary_item_list.append(item)
        else:
            pass
    cat_summary['items'] = cat_summary_item_list
    cat_summary['frequency'] = len(cat_summary_item_list)
    if len(cat_summary_item_list)==0:
        print("ALERT")
        break
    new_category_summarize_100.append(cat_summary)
```

```python
with open('data/category_summarize_100.json', 'w', encoding='utf-8') as f:
    f.write(json.dumps(new_category_summarize_100, ensure_ascii=False))
```
