```python
import torch
import torch.nn as nn
import torchvision.transforms as transforms
from PIL import Image
import os
```

```python
model = torch.hub.load('facebookresearch/dino:main', 'dino_vits16')
```

```python
model.eval()

# Define the image transformation pipeline
transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize((0.485, 0.456, 0.406), (0.229, 0.224, 0.225))
])
```

```python
import json
```

```python
for idx,image_file in enumerate(all_img_files):
  if idx%100==0:
    print(idx)
  image_path = 'polyvore-images-subset/'+image_file
  image = Image.open(image_path)
  if image.mode=='L':
    image = image.convert('RGB')
  image = transform(image)
  image = image.unsqueeze(0)

  # Pass the image through the model to get the image embedding
  embedding = model(image)

  # Print the shape of the image embedding tensor
  vector_name = image_file[:-4] + '.json'
  with open('polyvore_vit_image_vectors/' + vector_name, 'w') as f:
              f.write(json.dumps(embedding.tolist()[0]))

```
